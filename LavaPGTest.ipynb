```python
import policies
import scenarios
import torch as pt
import numpy as np

def sample_initial_dist():
    return np.random.uniform(0, 5)

def sample_sensor_noise(cov):
    return np.random.normal(0, np.sqrt(cov))

scenario = scenarios.LavaScenario(sample_initial_dist, lambda: sample_sensor_noise(0.001))
```

```python
np.random.seed(0)
pt.manual_seed(0)

#policy = policies.PGPolicy(scenario, 10, [64, 32], 0.1 * np.eye(scenario.ninputs))
#policy.train(nsamples=1000, training_iterations=200)

ntrvs = 2

policy = policies.MINEPolicy(scenario, 5, ntrvs, [256], 0.1 * np.eye(ntrvs), [256], 0.1 * np.eye(scenario.ninputs), 0.01, {'hidden_size' : 100})
policy.train(nsamples=500, training_iterations=700, qlr=0.0001, pilr=0.0001)
```

```python
def test_sample_initial_dist():
    return np.random.normal(2.5, 0.01)

def test_sample_sensor_noise(cov):
    return np.random.normal(0, np.sqrt(cov))

test_scenario = scenarios.LavaScenario(test_sample_initial_dist, lambda: test_sample_sensor_noise(1))

policy._scenario = test_scenario

states, outputs, trvs, inputs, costs = policy.rollout(10000)

costs.sum(axis=0).mean()
```
